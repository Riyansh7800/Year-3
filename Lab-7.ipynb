#### Consider the following data frame containing a family name, gender of the family member and her/his monthly income in each record


```python
import numpy as np
import pandas as pd
import random

FamilyDf = pd.DataFrame(columns=['Name', 'Gender', 'Monthly Income'])
FamilyDf['Name']=['Shah','Vats','Vats','Kumar','Vats','Kumar','Shah','Shah','Kumar','Vats']
FamilyDf['Gender']=['Male','Male','Female','Female','Female','Male','Male','Female','Female','Male']
FamilyDf['Monthly Income']=[114000.00, 65000.00, 43150.00, 69500.00, 155000.00, 103000.00, 55000.00, 112400.00,81030.00, 71900.00]
```

```python
FamilyDf
```

```python
# a. Calculate and display familywise gross monthly income.
Part1=FamilyDf.groupby(by=['Name']).mean(['Monthly Income'])
Part1
```

```python
#b Calculate and display the member with the highest monthly income in a family.
maxincome=FamilyDf.groupby(by=['Name']).max()
maxincome
```

```python
#c Calculate and display monthly income of all members with income greater than Rs. 60000.00.

part3=FamilyDf[FamilyDf['Monthly Income'] > 60000]
part3
```

```python
#d Calculate and display the average monthly income of the female members in the Shah family.
condition1= FamilyDf['Gender']=='Female' 
condition2=FamilyDf['Name']=='Shah'

# part4=FamilyDf[(FamilyDf['Name'] == 'Shah') & (FamilyDf['Gender'] == 'Female')]
part4=FamilyDf.where(condition1&condition2)
print(part4)
mean_income_shah_female = part4['Monthly Income'].mean()

print("Average income of shah family feamle",mean_income_shah_female)
```

#### Question 2 : Create a DataFrame DS with two columns 'Dates' and "Sale containing all dates of January 2021 and 31 random integers between 500 and 1000 respectively. Add another column 'Moving Avg to DS containing the rolling average of 5 consecutive values in the "Sale' column. Plot simple line plots between 'Dates' and 'Sale' as well as 'Dates' and 'Moving Avg'. Explain the utility of the rolling method with respect to these plots.

```python
import numpy as np
import pandas as pd
import random

dates = pd.date_range(start="2021-01-01", end="2021-01-31")
sales = np.random.randint(500, 1001, size=31)

DS = pd.DataFrame({'Dates': dates, 'Sale': sales})
DS
```

```python
DS['Moving Avg']=DS['Sale'].rolling(5).mean()
DS
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5)) 
plt.plot(DS['Dates'], DS['Sale'], marker='o', linestyle='--')
plt.title('Sales Over Time')
plt.xlabel('Dates')
plt.ylabel('Sale')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5)) 
plt.plot(DS['Dates'], DS['Sale'], marker='o', linestyle='--')
plt.plot(DS['Dates'], DS['Moving Avg'], marker='o', linestyle='-')
plt.title('Sales Over Time')
plt.xlabel('Dates')
plt.ylabel('Sale')
plt.tight_layout()
plt.show()
```
