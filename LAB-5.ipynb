```python
import numpy as np
import pandas as pd
data=np.random.randint(0,20,(50,3))
df=pd.DataFrame(data,columns=['A','B','C'],dtype=float)
df
```

```python
#Part1
import random
replace_n = int(50*3 * 0.10)
replace_i = random.sample(range(50*3), replace_n)

for i in replace_i:
    rowindex=i%50
    colindex=i%3
    df.iloc[rowindex,colindex]=None
      


df
```

```python
#Part-2
dnull=df.isnull().sum()
print(dnull)
```

```python
#Part-3
nulldf=df.dropna(axis=1,thresh=(len(df)-5))
nulldf
```

```python
#Part-4
#Sort the dataframe on the basis of the first column.

df.sort_values(by=['A'])
```

```python
#Remove all duplicates from the first column
df = df.drop_duplicates(subset=['A'])
df
```

```python
#Find the correlation between first and second column and covariance between second and third column.

corealtion=df['A'].corr(df['B'])
covariance = df['B'].cov(df['C'])
print("The Corelation is",corealtion)
print("The Covariance is",covariance)
```

```python
#Discretize second column and create 5 bins
df['B'].fillna(0, inplace=True)
discertize=pd.cut(df['B'],bins=5)
print(discertize)
```
