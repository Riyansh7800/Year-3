## Question-1: Consider the following two data frames: Personnel-Details & Education-details:
### a. Make Name as the row index of the both Data frames and Combine these two DFs column-wise
### b Find total number of individuals who are postgraduate and whose details are given
### c For each name starting with same letter, arrange the data frame
### d For each group (first letter of the name, age) Find average salary of the group

```python
import numpy as np
import pandas as pd
import random
```

```python
PersonnelDetails=pd.DataFrame(columns=["Name","Age","Salary","Occupation"])
PersonnelDetails["Name"]=['A','B','C','A1','A2']
PersonnelDetails['Age']=[20,25,45,48,20]
PersonnelDetails["Salary"]=[10000,12000,14000,10000,8000]
PersonnelDetails["Occupation"]=["Service","Service","Buisness","Buisness",np.nan]
PersonnelDetails
```

```python
EducationDetails=pd.DataFrame(columns=["Name","Intermediate","Graduate","PostGraduate"])
EducationDetails["Name"]=['A','B','C','A1','DD']
EducationDetails["Intermediate"]=["Yes","Yes","Yes","No","Yes"]
EducationDetails["Graduate"]=["Yes","No","Yes",np.nan,"Yes"]
EducationDetails["PostGraduate"]=["Yes",np.nan,"Yes","No","No"]
EducationDetails
```

```python
#a. Make Name as the row index of the both Data frames and Combine these two DFs column-wise
PersonnelDetails.set_index("Name",inplace=True)
EducationDetails.set_index("Name",inplace=True)
print(EducationDetails)
print(PersonnelDetails)
```

```python
part1=PersonnelDetails.join(EducationDetails,how='outer')
part1
```

```python
#b Find total number of individuals who are postgraduate and whose details are given
#Here i have assumed that the two dfs needs to be merged together and then we will execute the query
condition=part1['PostGraduate']=="Yes"
condition2=part1.notna().all(axis=1)
result=part1.where(condition & condition2).dropna()

print("Total number of postgraduate individuals with non-null details:", len(result))
```

```python
#c For each name starting with same letter, arrange the data frame
part_c = part1.sort_values(by='Name')
part_c
```

```python
#d For each group (first letter of the name, age) Find average salary of the group
part4= part1.groupby([part1.index.str[0], 'Age'])
avg_sal = part4['Salary'].mean()
avg_sal.columns = ['FirstLetter', 'Age', 'AverageSalary']
avg_sal
```

### FamilyName Gender Monthly Income (Rs.)

### Vats Male 65000.00
### Vats Female 43150.00
### Kumar Female 66500.00
### Vats Female 255000.00
### Kumar Male 103000.00
### Shah Male 55000.00
### Shah Female 112400.00
### Kumar Female 81030.00
### Vats Male 71900.00

```python
#Creatig DF

FamilyDf = pd.DataFrame(columns=['Name', 'Gender', 'Monthly Income'])
FamilyDf['Name']=['Shah','Vats','Vats','Kumar','Vats','Kumar','Shah','Shah','Kumar','Vats']
FamilyDf['Gender']=['Male','Male','Female','Female','Female','Male','Male','Female','Female','Male']
FamilyDf['Monthly Income']=[114000.00, 65000.00, 43150.00, 69500.00, 155000.00, 103000.00, 55000.00, 112400.00,81030.00, 71900.00]
```

```python
FamilyDf
```

```python
# a. Calculate and display familywise gross monthly income.
Part1=FamilyDf.groupby(by=['Name'])['Monthly Income'].sum()
Part1
```

```python
#Display the highest and lowest monthly income for each family name
result = FamilyDf.groupby('Name')['Monthly Income'].agg(['max', 'min'])
result.columns = ['Highest Income', 'Lowest Income']
result
```

```python
#c Calculate and display monthly income of all members with income less than Rs. 80000.00.

part3=FamilyDf[FamilyDf['Monthly Income'] < 80000]
part3
```

```python
#d Calculate and display the number of  female members and thier average income
condition1= FamilyDf['Gender']=='Female' 
part4=FamilyDf.where(condition1)
print("The number of Female members are ",len(part4.value_counts()))
print("Their mean income is",part4['Monthly Income'].mean())
```

```python
#Delete rows with Monthly income less than the average income of all members
average_income=FamilyDf['Monthly Income'].mean()
part5=FamilyDf.where(FamilyDf['Monthly Income']>average_income)
part5.dropna(inplace=True)
part5
```

```python
#Import the data into two dataframes
day1=pd.read_excel('../DAV-SEM5/Data/workshop_day1.xlsx')
day2=pd.read_excel('../DAV-SEM5/Data/workshop_day2.xlsx')
print(day1)
print(day2)
```

```python
day2
```

```python
#f. Perform merging of the two dataframes to find the names of students who had attended the workshop on both days.
part_a=day1.merge(day2,on='Name')
part_a['Name']
```

```python
#Find names of all students who have attended workshop on either of the days.
part_b=day1.merge(day2, how='outer') #We can also use concat method
names=part_b['Name'].unique()
print("Names of all students who have attended workshop on either of the days.",'\n',names)
```

```python
#Merge two data frames row-wise and find the total number of records in the data frame
#to merge row wise we can use pd.concat
part_c=pd.concat([day1,day2],axis=0)
print("The total number of records in the data frame is ", len(part_c))
```

```python
# Merge two data frames and use two columns names and duration as multi-row indexes.
# Generate descriptive statistics for this multi-index where feature is categorical attribute.
part_d=day1.merge(day2,on=["Name","Duration"],how='outer')
# part_d.set_index(['Name','Duration'],inplace=True)
# part_d.describe()
part_d
```
