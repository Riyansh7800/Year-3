```python
import numpy as np

int_array=np.array([10,12,4,6,3,18])

output_array=np.argsort(int_array)
print("Array is: ",int_array)
print("Argsort returns the indices of elements in original array sorted in ascending order:",'\n',output_array)

out2=np.sort(int_array)
print("Sort returns the  elements sorted in ascending order:",'\n',out2)
```
