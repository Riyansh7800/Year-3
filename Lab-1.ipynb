 Question: 1 
 Given below is a dictionary having two keys ‘Boys’ and ‘Girls’ and having two lists of heights of five Boys and Five Girls respectively as values associated with these keys Original dictionary of lists: {'Boys': [72, 68, 70, 69, 74], 'Girls': [63, 65, 69, 62, 61]} From the given dictionary of lists create the following list of dictionaries: [{'Boys': 72, 'Girls': 63}, {'Boys': 68, 'Girls': 65}, {'Boys': 70, 'Girls': 69}, {'Boys': 69, 'Girls': 62}, {‘Boys’:74, ‘Girls’:61]

```python
original_dict={'Boys': [72, 68, 70, 69, 74], 'Girls': [63, 65, 69, 62, 61]}
list1=list()
for i in range(len(original_dict['Girls'])):
    dictionary=dict()
    dictionary['Boys']=original_dict['Boys'][i]
    dictionary['Girls']=original_dict['Girls'][i]
    list1.append(dictionary)

print(list1)
```

Question: 2 :
Given are three lists where L1 has names of n students, L2 has marks and L3 has hobbies of n students.Using three lists, create the following dictionary with hobbies as keys and (names,marks) as values. In case more than one student has same hobby then values must be appended for the same key instead of overwriting.E.g.
L1=['A','B','C']   L2=[100,40,50] L3=['painting','music','painting']  
then output should be:
dict1={painting:(('A',100),('C',50)), 'music':('B',40)}

```python
L1=['A','B','C']
L2=[100,40,50]
L3=['painting','music','painting']

dict1=dict()
for i in range(len(L3)):
    name=L1[i]
    marks=L2[i]
    hobby=L3[i]

    if hobby not in dict1:
        dict1[hobby]=[]

    dict1[hobby].append((name,marks))

for hobby,values in dict1.items():
        dict1[hobby]=tuple(values)

print(dict1)
```

Question-3
Write a lambda function to arrange a list of names on the last letter of the name then print the sorted list on last letter.

```python
word_list = ["apple", "banana", "cherry", "date", "fig", "grape","orange"]

sorted_list = sorted(word_list, key=lambda word: word[-1])

print(sorted_list)
```

Question-4: Make groups in a list of names on the basis of last name and store in a
dictionary using itertools


```python
import itertools as it

names = [ "Julie Kapoor","Aryan Singh", "Vriddhi Singh", "Evanya Chopra", "Priyanka Chopra"]

name_groups = {}
for last_name, group in it.groupby(names, key=lambda name: name.split()[-1]):
    name_groups[last_name] = list(group)

print(name_groups)
```

```python
import numpy as np

array = np.random.randint(0, 10, size=(3, 5))

mean = np.mean(array, axis=1)
std_dev = np.std(array, axis=1)
variance = np.var(array, axis=1)


print("Original Array:")
print(array)
print("Mean along the second axis:")
print(mean)
print("Standard Deviation along the second axis:")
print(std_dev)
print("Variance along the second axis:")
print(variance)

```

```python
B = np.array([56, 48, 22, 41, 78, 91, 24, 46, 8, 33])
sorted_indices = np.argsort(B)

print("Original Array:")
print(B)
print("Indices of Sorted Elements:")
print(sorted_indices)
```

```python
m = int(input("Enter the number of rows (m): "))
n = int(input("Enter the number of columns (n): "))

# Create a random m x n array of integers
array = np.random.randint(0, 100, size=(m, n))

print("Original Array:")
print(array)
print("Shape:", array.shape)
print("Type:", type(array))
print("Data Type:", array.dtype)

# Reshape it into an nxm array
reshaped_array = array.reshape(n, m)
print("Reshaped Array:")
print(reshaped_array)
```

```python

# Create a sample array with some zeros, non-zeros, and NaN values
array = np.array([1.0, 0.0, 2.0, 0.0, np.nan, 3.0])

# Test for zero elements and record their indices
zero_indices = np.where(array == 0.0)[0]

# Test for non-zero elements and record their indices
nonzero_indices = np.where(array != 0.0)[0]

# Test for NaN elements and record their indices
nan_indices = np.where(np.isnan(array))[0]

print("Original Array:")
print(array)
print("Indices of Zero Elements:")
print(zero_indices)
print("Indices of Non-Zero Elements:")
print(nonzero_indices)
print("Indices of NaN Elements:")
print(nan_indices)
```

Question-7: 
Generate a 2D array having values from 1 to 100 of shape (4,5).
I. Multiply all elements that are greater than 50 by 2 
II. Extract 2x2 lower-right matrix from the given array and find its largest element

```python
array = np.random.randint(1,101 , size=(4, 5))
print("Original Array : ")
print(array)


# I. Multiply all elements greater than 50 by 2
indices = np.where(array > 50)
array[indices] *= 2

print("Array after Multiplication: ")
print(array)
```

```python
# II. Extract the 2x2 lower-right matrix and find its largest element
lower_right_matrix = array[-2:, -2:]  # Extract the lower-right 2x2 matrix
largest_element = np.max(lower_right_matrix)

print("\nLower-Right 2x2 Matrix:")
print(lower_right_matrix)
print("\nLargest Element in Lower-Right Matrix:", largest_element)
```
